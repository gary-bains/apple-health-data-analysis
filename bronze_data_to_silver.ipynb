```python
import pandas as pd
import json
```

```python
# Load the line sperated JSON into an Panada Data Frame

bronze_data = []

with open('bronze_data/records.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        strippedLine = line.strip()
        if strippedLine:
            bronze_data.append(json.loads(strippedLine))

bronze_df = pd.DataFrame(bronze_data)

bronze_df
```

```python
# Since we only need type 'type', 'unit', 'creationDate' and 'value'
# and don't care about 'sourceName', 'sourceVersion', 'device', 'startDate' and 'endDate' columns we can remove them

bronze_df_v2 = bronze_df[['type', 'unit', 'creationDate', 'value']]
bronze_df_v2
```

```python
bronze_df_v2.iloc[0].unit
```

```python
types = bronze_df_v2['type'].unique()
types
```

```python
for data_type in types:
    data_type_df = bronze_df_v2[bronze_df_v2['type'] == data_type]

    # only keeping the data / removing redundant info
    data_type_df_v2 = data_type_df[['creationDate', 'value']]

    data_type_name = data_type.replace('HKQuantityTypeIdentifier', '')
    new_col_name = f'{data_type_name} in {data_type_df.iloc[0].unit}'

    data_type_df_v3 = data_type_df_v2.rename(columns={'value': new_col_name})
    data_type_df_v3.to_csv(f'silver_data/{data_type_name}.csv', sep=',', index=False)

    print(data_type_df_v3.head())
    print(data_type_df_v3.shape[0], 'rows')
    print()
```
